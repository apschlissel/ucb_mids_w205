# Allison's Report and Recommendations - SF Trip Data

Using SF Bike Trip Data, I compiled the most popular trip data, along with recommendations to further enhance our sales.

## What are the 5 most popular trips that you would call "commuter trips"?
The five most popular trips are:
1. San Francisco Caltrain 2 (330 Townsend) TO Townsend at 7th
2. 2nd at Townsend TO Harry Bridges Plaza (Ferry Building)
3. Townsend at 7th TO San Francisco Caltrain 2 (330 Townsend)
4. Harry Bridges Plaza (Ferry Building) TO 2nd at Townsend
5. Embarcadero at Folsom TO San Francisco Caltrain (Townsend at 4th)

Assumptions: 
* commuters are subscribers
* commute Monday through Friday

I filtered my query on subscribers and on the days Monday through Friday. I then got the 5 most popular trips

```python
%%bigquery df_commute

SELECT start_station_id, start_station_name, end_station_id, end_station_name, COUNT(*) as count
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE subscriber_type = 'Subscriber' 
AND EXTRACT(DAYOFWEEK FROM start_date) > 0 and EXTRACT(DAYOFWEEK FROM start_date) < 6
GROUP BY start_station_id, start_station_name, end_station_id, end_station_name
ORDER BY COUNT DESC
LIMIT 20
```

```python
df_commute
```

```python
df_commute['combined_trip'] = df_commute.start_station_name + ' & ' + df_commute.end_station_name
df_commute.plot.bar(x='combined_trip', y='count')
```

## Supporting Data

```python
%%bigquery df_subscriber

SELECT start_station_id, start_station_name, end_station_id, end_station_name, COUNT(*) as count
FROM `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE subscriber_type = 'Subscriber'
GROUP BY start_station_id, start_station_name, end_station_id, end_station_name
ORDER BY COUNT DESC
LIMIT 20
```

```python
#most of the subscriber top trips are in the SOMA/FiDi neighborhoods of SF
df_subscriber
```

```python
%%bigquery df_gender

SELECT member_gender, count(*) as count
FROM `bigquery-public-data.san_francisco_bikeshare.bikeshare_trips`
GROUP BY member_gender
```

```python
#takeaway: the largest category of members is Null, followed by male 
df_gender.plot.bar(x='member_gender', y='count')
```

```python
%%bigquery df_birthyear

SELECT member_birth_year, count(*) as count
FROM `bigquery-public-data.san_francisco_bikeshare.bikeshare_trips`
GROUP BY member_birth_year
ORDER BY count DESC 
LIMIT 10
```

```python
#The top given ages are all in the Millennial age group
df_birthyear.plot.bar(x='member_birth_year', y='count')
```
